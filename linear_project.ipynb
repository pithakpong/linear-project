# pre process image

resize image using bilinear interpolation

```python

#for backward resizing
def custom_resize_array(input_array, new_width, new_height):
    # Get the original array dimensions
    height, width = input_array.shape

    # Calculate scaling factors
    x_scale = width / new_width
    y_scale = height / new_height

    # Create an empty array for the resized output
    resized_array = np.zeros((new_height, new_width), dtype=input_array.dtype)

    # Iterate through the new array dimensions and apply bilinear interpolation
    for y in range(new_height):
        for x in range(new_width):
            src_x = x * x_scale
            src_y = y * y_scale

            x0, y0 = int(src_x), int(src_y)
            x1, y1 = min(x0 + 1, width - 1), min(y0 + 1, height - 1)

            dx = src_x - x0
            dy = src_y - y0

            # Bilinear interpolation
            top_left = input_array[y0, x0] * (1 - dx) * (1 - dy)
            top_right = input_array[y0, x1] * dx * (1 - dy)
            bottom_left = input_array[y1, x0] * (1 - dx) * dy
            bottom_right = input_array[y1, x1] * dx * dy

            resized_array[y, x] = top_left + top_right + bottom_left + bottom_right

    return resized_array
```

```python
from keras.preprocessing.image import ImageDataGenerator
from keras.layers import Convolution2D, MaxPooling2D
from keras.layers import Activation, Dropout, Flatten, Dense, BatchNormalization,MaxPooling2D
from keras.models import Sequential
from keras.models import model_from_json
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping,ReduceLROnPlateau
from tensorflow.keras.losses import CategoricalCrossentropy
import matplotlib.pyplot as plt
import matplotlib
import brewer2mpl
import numpy as np
import pandas as pd
```

```python
def create_model():
    model = Sequential([
        Conv2D(filters=128, kernel_size=(5, 5), padding='valid', input_shape=(48 ,48 , 1)),
        Activation('relu'),
        MaxPooling2D(pool_size=(2, 2)),
        BatchNormalization(),

        Conv2D(filters=64, kernel_size=(3, 3), padding='valid', kernel_regularizer=l2(0.00005)),
        Activation('relu'),
        MaxPooling2D(pool_size=(2, 2)),
        BatchNormalization(),

        Conv2D(filters=32, kernel_size=(3, 3), padding='valid', kernel_regularizer=l2(0.00005)),
        Activation('relu'),
        MaxPooling2D(pool_size=(2, 2)),
        BatchNormalization(),

        Flatten(),

        Dense(units=256, activation='relu'),
        Dropout(0.3),
        Dense(units=6, activation='softmax')
    ])

    return model
```

```python
import os
import random
import shutil
dataset_dir = "dataset"
validation_ratio = 0.3
validation_dir = os.path.join(dataset_dir, "validation")

if not os.path.exists(validation_dir):
    os.makedirs(validation_dir)
for emotion in ["angry", "sad", "happy","fear","neutral","surprise"]:
    train_emotion_dir = os.path.join(dataset_dir, "train", emotion)
    validation_emotion_dir = os.path.join(validation_dir, emotion)

    if not os.path.exists(validation_emotion_dir):
        os.makedirs(validation_emotion_dir)

    # List all files in the train emotion folder
    files = os.listdir(train_emotion_dir)

    # Calculate the number of files to move to the validation set
    num_validation_files = int(validation_ratio * len(files))

    # Randomly select files to move
    validation_files = random.sample(files, num_validation_files)

    # Move the selected files to the validation set
    for file in validation_files:
        src_path = os.path.join(train_emotion_dir, file)
        dest_path = os.path.join(validation_emotion_dir, file)
        shutil.move(src_path, dest_path)
```

```python
import os
dataset_dir = "dataset"

train_dir = os.path.join(dataset_dir, "train")
validation_dir = os.path.join(dataset_dir, "validation")
test_dir = os.path.join(dataset_dir, "test")

# Create a dictionary to map class names to label indices
class_mapping = {
    "angry": 0,
    "fear": 1,
    "happy": 2,
    "neutral":3,
    "sad":4,
    "surprise":5
    # Add more classes as needed
}

# Initialize empty label arrays
def label_data(data_dir):
    x_data = []  # Store the image data (file paths)
    y_data = []  # Store the corresponding labels

    for class_name in os.listdir(data_dir):
        class_dir = os.path.join(data_dir, class_name)
        if os.path.isdir(class_dir):
            class_label = class_mapping.get(class_name, -1)  # Assign a label based on the dictionary
            if class_label != -1:
                for filename in os.listdir(class_dir):
                    image_path = os.path.join(class_dir, filename)
                    image = cv2.imread(image_path)
                    image = np.array(image).astype(np.float32)/255.0
                    x_data.append(image)
                    y_data.append(class_label)

    return x_data, y_data
```

```python
import cv2
x_train, y_train = label_data(os.path.join(dataset_dir, "train"))
x_validation, y_validation = label_data(os.path.join(dataset_dir, "validation"))
x_test, y_test = label_data(os.path.join(dataset_dir, "test"))
```

```python
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam
x_train = np.array(x_train)
y_train = np.array(y_train)
x_validation = np.array(x_validation)

y_validation_encoded = to_categorical(y_validation, 6)
y_train_encoded = to_categorical(y_train, 6)
print(x_train.shape,y_train_encoded.shape)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=np.sqrt(0.1), patience=5)
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss=CategoricalCrossentropy(), metrics=['accuracy'])
history = model.fit(x_train, y_train_encoded, batch_size=32, epochs=50,validation_data=(x_validation, y_validation_encoded), callbacks=[reduce_lr])
```

```python
import matplotlib.pyplot as plt

# Access the training history
training_loss = history.history['loss']
validation_loss = history.history['val_loss']
training_accuracy = history.history['accuracy']
validation_accuracy = history.history['val_accuracy']
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(training_loss, label='Training Loss')
plt.plot(validation_loss, label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.subplot(1, 2, 2)
plt.plot(training_accuracy, label='Training Accuracy')
plt.plot(validation_accuracy, label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.show()
```

#analysis between dense layer and dense_1 layer how similar feature extraction using cosine similarity

```python
from tensorflow.keras.preprocessing import image
import numpy as np
features = model.predict(x_test[:100])

# Calculate cosine similarity between feature vectors
cosine_similarities = []
for i in range(len(features)):
    for j in range(i + 1, len(features)):
        similarity = np.dot(features[i].flatten(), features[j].flatten()) / (np.linalg.norm(features[i]) * np.linalg.norm(features[j]))
        cosine_similarities.append(similarity)
print(len(cosine_similarities))
# Visualize or print the cosine similarities with layer names
for similarity in cosine_similarities:
    layer1_name = model.layers[-3].name
    layer2_name = model.layers[-1].name
    print(f"Cosine Similarity between {layer1_name} and {layer2_name}: {similarity:.2f}")
```
